### 蒙特卡罗方法计算π值

```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
def mc_pi(num):
    # 随机生成指定数量的点     
    x = np.random.random(num)
    y = np.random.random(num)
    
    inside_unit_circle = x ** 2 + y ** 2 < 1
    
    unit_circle_x = np.linspace(0, 1, 1000)
    unit_circle = [unit_circle_x, np.sqrt(1 - unit_circle_x ** 2)]
    plt.figure(figsize=(10, 10))
    plt.plot(*unit_circle, color='black')
    plt.scatter(x[inside_unit_circle], y[inside_unit_circle], c='green', s=1,)
    plt.scatter(x[~inside_unit_circle], y[~inside_unit_circle], c='red', s=1)
    
    plt.title('Value of $ \pi $ = ' + str(4.0 * np.sum(inside_unit_circle)/float(num)))

mc_pi(100000)
```
